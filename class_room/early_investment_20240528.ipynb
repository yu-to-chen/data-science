```python
import numpy as np
import matplotlib.pyplot as plt

# Define the future value calculation function
def future_value_annuity(P, r, n):
    return P * (((1 + r) ** n - 1) / r)

# Parameters for John and Mary
john_annual_contribution_revised = 6000
mary_annual_contribution = 10000
r = 0.07

# Calculate John's investment value at age 45
john_FV_45 = future_value_annuity(john_annual_contribution_revised, r, 20)

# Calculate John's investment value at age 65
john_FV_65 = john_FV_45 * (1 + r) ** 20

# Calculate Mary's investment value at age 65
mary_FV_65 = future_value_annuity(mary_annual_contribution, r, 30)

# Define the years for John and Mary
years = np.arange(25, 66)

# Calculate the cumulative investment for John and Mary
john_cumulative_investment_revised = [john_annual_contribution_revised * (year - 25 + 1) if year <= 45 else john_annual_contribution_revised * (45 - 25 + 1) for year in years]
mary_cumulative_investment = [mary_annual_contribution * (year - 35 + 1) if year >= 35 else 0 for year in years]

# Calculate the investment value for each year John is investing
john_investment_value_revised = [future_value_annuity(john_annual_contribution_revised, r, year - 25) if year <= 45 else future_value_annuity(john_annual_contribution_revised, r, 20) * (1 + r) ** (year - 45) for year in years]

# Calculate the investment value for each year Mary is investing
mary_investment_value = [future_value_annuity(mary_annual_contribution, r, year - 35) if year >= 35 else 0 for year in years]

# Create a plot of the investment growth over time
plt.figure(figsize=(10, 6))
plt.plot(years, john_investment_value_revised, label='John (Starts at 25, Stops at 45)', color='blue')
plt.plot(years, mary_investment_value, label='Mary (Starts at 35, Stops at 65)', color='orange')
plt.xlabel('Age')
plt.ylabel('Investment Value ($)')
plt.title('Revised Investment Growth: John vs. Mary')
plt.legend()
plt.grid(True)
plt.show()

# Display the final values
john_FV_65, mary_FV_65
```
